```python
# First, let's make a small helper function to make sure we get a real number
# We MUST define the function *before* we can use it.
def get_number(prompt):
    """Keeps asking the user until they enter a valid number."""
    while True:
        try:
            value = float(input(prompt))
            return value
        except ValueError:
            print("Invalid input. Please enter a valid number.")

# --- Main Calculator Loop ---
# This will keep running until the user decides to exit
while True:

    print("\n" + "="*30)
    print("  Welcome to the Python Calculator! ")
    print("="*30)
    
    # Get the two numbers for the calculation      
    num1 = get_number("Enter first number: ")
    num2 = get_number("Enter second number: ")

    # Get the operation from the user
    # Now includes the "5. Exit" option as requested
    print("\nSelect an operation:")
    print("  1. Addition")
    print("  2. Subtraction")
    print("  3. Multiplication")
    print("  4. Division")
    print("  5. Exit") # This is from the PDF example

    operation = input("\nEnter your choice (1/2/3/4/5): ")
    
    # This check now looks for '5' instead of 'q'
    if operation not in ['1', '2', '3', '4', '5']:
        print("Invalid choice. Please enter a number between 1 and 5.")
        continue # This skips the rest of the loop and starts from the top
    
    # --- Perform the calculation based on the operation ---
    
    if operation == '1':
        result = num1 + num2
        op_symbol = '+'
        print(f"The result of {num1} {op_symbol} {num2} = {result}")

    elif operation == '2':
        result = num1 - num2
        op_symbol = '-'
        print(f"The result of {num1} {op_symbol} {num2} = {result}")

    elif operation == '3':
        result = num1 * num2
        op_symbol = '*'
        print(f"The result of {num1} {op_symbol} {num2} = {result}")

    elif operation == '4':
        if num2 == 0:
            print("Error! Division by zero is not allowed.")
        else:
            result = num1 / num2
            op_symbol = '/' 
            print(f"The result of {num1} {op_symbol} {num2} = {result}")
    
    elif operation == '5':
        break # This stops the 'while True' loop

    # --- This is the "another calculation" loop from the PDF ---
    # It will only run if the user did *not* pick '5'
    
    # We need a new loop here to make sure they type 'yes' or 'no'
    while True:
        another = input("\nDo you want to perform another calculation? (yes/no): ").lower().strip()
        
        if another == 'yes':
            break # This breaks the *inner* loop and lets the main loop run again
        elif another == 'no':
            break # This breaks the *inner* loop
        else:
            print("Invalid input. Please enter 'yes' or 'no'.")
    
    # Now we check the answer. If it was 'no', we break the *main* loop to exit.
    if another == 'no':
        break

# This final message will print only when the loop breaks
print("\nThank you for using the calculator. Goodbye!")


```
